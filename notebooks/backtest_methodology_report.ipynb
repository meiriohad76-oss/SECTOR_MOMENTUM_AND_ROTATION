````python
from pathlib import Path
import json
import pandas as pd

ROOT = Path.cwd()
REPORT = ROOT / "docs/backtest_report.md"
METHODOLOGY_REPORT = ROOT / "docs/backtest_methodology_report.md"
EQUITY = ROOT / "docs/backtest_equity.csv"
METADATA = ROOT / "docs/backtest_metadata.json"

metadata = json.loads(METADATA.read_text(encoding="utf-8"))
equity = pd.read_csv(EQUITY, index_col="date", parse_dates=True)
metadata
````

````python
normalized_equity = equity / equity.iloc[0]
drawdown = equity / equity.cummax() - 1.0

display(normalized_equity.tail())
display(drawdown.tail())
````
